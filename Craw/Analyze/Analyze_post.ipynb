## Ng√†y trong tu·∫ßn ƒë·ªÉ ƒëƒÉng b√†i

```python
# Chuy·ªÉn c·ªôt 'time' sang ki·ªÉu datetime ƒë·ªÉ chu·∫©n h√≥a ƒë·ªãnh d·∫°ng d·ªØ li·ªáu th·ªùi gian
df['time'] = pd.to_datetime(df['time'])

# √Ånh x·∫° ng√†y trong c·ªôt 'time' sang t√™n c·ªßa ng√†y trong tu·∫ßn (Monday, Tuesday, ...) v√† l∆∞u k·∫øt qu·∫£ v√†o c·ªôt 'week_day'
week_day = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['week_day'] = df['time'].apply(lambda x: week_day[x.weekday()])
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# T·∫°o m·∫£ng m√†u s·∫Øc t·ª´ d√£y m√†u rainbow
colors = cm.rainbow(np.linspace(0, 1, len(df['week_day'].unique())))

# V·∫Ω bi·ªÉu ƒë·ªì c·ªôt ƒë·ªÉ th·ªÉ hi·ªán s·ªë l∆∞·ª£ng b√†i ƒëƒÉng theo ng√†y trong tu·∫ßn
ax = df['week_day'].value_counts().plot(kind='bar', figsize=(16,8), color=colors)
plt.title('S·ªë l∆∞·ª£ng b√†i ƒëƒÉng theo c√°c th·ª©')
plt.xlabel('Th·ª©')
plt.ylabel('S·ªë b√†i ƒëƒÉng')

# Xoay m·ªëc c·ªßa c·ªôt x theo chi·ªÅu ngang
plt.xticks(rotation=0, ha="right")

plt.show()
```

Emoji trong b√†i vi·∫øt

```python
import pandas as pd
import emoji

# Chuy·ªÉn c·ªôt text th√†nh d·∫°ng str
df['text'] = df['text'].astype(str)

# T·∫°o h√†m t√≠nh t·ªïng s·ªë emoji
def count_emoji(text):
    return emoji.emoji_count(text)

# Th√™m c·ªôt m·ªõi ƒë·ªÉ t√≠nh t·ªïng s·ªë emoji
df['emoji_count'] = df['text'].apply(count_emoji)

sum_emoji = df[df['emoji_count'] > 0]
print("C√≥",df['emoji_count'].sum(),"emoji ƒë∆∞·ª£c s·ª≠ d·ª•ng trong 150 b√†i vi·∫øt")

row_count = sum_emoji.shape[0]
print("Trong ƒë√≥ c√≥", row_count,"b√†i vi·∫øt s·ª≠ d·ª•ng emoji")
```
