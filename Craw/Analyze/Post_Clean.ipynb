```python
# Ch·ªçn c√°c c·ªôt c·∫ßn gi·ªØ l·∫°i
selected_columns = [
    "post_id",
    "text",
    "post_text",
    "time",
    "image",
    "video",
    "comments",
    "shares",
    "comments_full",
    "reactors",
    "reactions",
    "reaction_count"
]

# T·∫°o DataFrame m·ªõi t·ª´ c√°c c·ªôt ƒë√£ ch·ªçn
df_clean = df[selected_columns]
```

```python
df_clean
```

```python
df_clean['post_text'] = df_clean['post_text'].fillna('')

# Thay th·∫ø gi√° tr·ªã NaN b·∫±ng 0
df_clean['image'] = df_clean['image'].fillna(0)

# Thay th·∫ø gi√° tr·ªã kh√°c NaN b·∫±ng 1
df_clean['image'] = (df_clean['image'] != 0).astype(int)

# Thay th·∫ø gi√° tr·ªã NaN b·∫±ng 0
df_clean['video'] = df_clean['video'].fillna(0)

# Thay th·∫ø gi√° tr·ªã kh√°c NaN b·∫±ng 1
df_clean['video'] = (df_clean['video'] != 0).astype(int)
```
